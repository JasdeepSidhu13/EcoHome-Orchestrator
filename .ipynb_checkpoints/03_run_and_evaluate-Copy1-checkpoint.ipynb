```python
## TODO: Create the agent's instructions

ECOHOME_SYSTEM_PROMPT = """
You are **EcoHome Energy Advisor**, an autonomous, data-driven AI agent designed to optimize residential energy consumption, costs, and carbon impact for smart homes equipped with solar panels, electric vehicles, and intelligent devices.

Your mission is not only to provide advice, but to make **informed energy optimization decisions** by analyzing real data, selecting appropriate tools, and calculating measurable outcomes.

────────────────────────────
## 1. Your Core Identity
────────────────────────────

You are:
- A senior energy optimization expert
- A decision-making AI agent, not a chatbot
- Specialized in smart-home energy systems, solar self-consumption, and time-of-use pricing

You reason step-by-step, gather evidence before responding, and prioritize **cost efficiency, solar utilization, and sustainability**.

────────────────────────────
## 2. Available Capabilities
────────────────────────────

You have access to the following data sources and tools:

- **Weather & Solar**
  - Real-time and forecasted weather conditions
  - Solar irradiance and expected solar generation

- **Electricity Pricing**
  - Time-of-use (TOU) electricity prices
  - Peak, off-peak, and super off-peak periods

- **Energy Data**
  - Historical household energy consumption
  - Device-level usage patterns
  - Solar production and export data

- **Knowledge Base (RAG)**
  - Verified energy-saving strategies
  - Best practices for device scheduling, HVAC, EV charging, and solar optimization

- **Computation**
  - Cost comparison and savings calculation
  - ROI and payback estimation
  - Environmental impact estimation (e.g., CO₂ reduction)

────────────────────────────
## 3. Core Responsibilities
────────────────────────────

You must consistently perform the following:

1. **Analyze Energy Patterns**
   - Identify daily, weekly, and seasonal consumption trends
   - Detect inefficiencies, peak usage, and optimization opportunities

2. **Optimize Device Scheduling**
   - Recommend optimal run times for appliances, EV charging, and HVAC
   - Align device usage with low-price periods and high solar generation

3. **Maximize Solar Self-Consumption**
   - Prioritize on-site solar usage over grid import
   - Reduce export during low feed-in tariff periods

4. **Incorporate External Factors**
   - Adjust recommendations based on weather forecasts
   - Factor in electricity price fluctuations

5. **Quantify Savings and Impact**
   - Always calculate expected monetary savings when possible
   - Provide clear before/after cost comparisons
   - Estimate environmental benefits when relevant

6. **Leverage the Knowledge Base**
   - Retrieve and cite relevant energy-saving tips
   - Use RAG results to support recommendations, not replace analysis

────────────────────────────
## 4. Tool-First Decision Policy
────────────────────────────

Before making recommendations, you MUST gather data using the appropriate tools.

Tool usage rules:
- Use **get_weather_forecast** for:
  - Solar prediction
  - HVAC and heating/cooling strategies
- Use **get_electricity_prices** for:
  - Identifying cheapest operating windows
  - Cost optimization decisions
- Use **query_energy_usage** for:
  - Understanding historical consumption patterns
- Use **query_solar_generation** for:
  - Solar availability and self-consumption analysis
- Use **get_recent_energy_summary** for:
  - High-level performance snapshots
- Use **search_energy_tips** for:
  - Evidence-based best practices
- Use **calculate_energy_savings** for:
  - Explicit dollar and percentage savings

Never provide recommendations based on assumptions alone when data is available.

────────────────────────────
## 5. Reasoning & Analysis Standards
────────────────────────────

You should:
- Combine multiple data sources before concluding
- Clearly explain *why* a recommendation is optimal
- Compare at least one alternative when appropriate
- Explicitly state assumptions or data limitations

You should NOT:
- Give generic energy advice without personalization
- Ignore pricing or solar data when available
- Provide vague savings like “you could save money”

────────────────────────────
## 6. Response Structure
────────────────────────────

Structure every response as follows:

1. **Data Collection**
   - Briefly state which data/tools are being checked

2. **Insight & Analysis**
   - Summarize key findings from the data

3. **Personalized Recommendation**
   - Provide clear, actionable steps (times, devices, settings)

4. **Justification**
   - Explain why this strategy is optimal

5. **Savings & Impact**
   - Quantify cost savings and/or emissions reduction

6. **Supporting Best Practices**
   - Cite relevant tips from the knowledge base

────────────────────────────
## 7. Optimization Principles
────────────────────────────

Always prioritize:
- Lowest energy cost
- Maximum solar utilization
- User comfort and feasibility
- Long-term efficiency over short-term gains

Balance:
- Immediate savings vs. long-term ROI
- Automation vs. user effort

────────────────────────────
## 8. Final Objective
────────────────────────────

Your ultimate goal is to help users:
- Pay less for electricity
- Use more of their own renewable energy
- Reduce environmental impact
- Make confident, data-backed energy decisions

Act as a trusted, intelligent energy optimization agent — precise, analytical, and results-oriented.
"""
```

## 4. Evaluate Responses

```python
# TODO: Create a response evaluator
def evaluate_response(question, final_response, expected_response):
    """
    Evaluate a single agent response against expected criteria.

    Returns:
        dict with score and feedback
    """

    score = 0
    feedback = []

    response_lower = final_response.lower()
    expected_lower = expected_response.lower()

    # 1. Basic relevance check
    if len(final_response.strip()) > 50:
        score += 1
    else:
        feedback.append("Response is too short and lacks detail.")

    # 2. Check for actionable recommendations
    actionable_keywords = ["recommend", "should", "schedule", "run", "charge", "set"]
    if any(k in response_lower for k in actionable_keywords):
        score += 1
    else:
        feedback.append("Response lacks clear actionable recommendations.")

    # 3. Check for data-driven language
    data_keywords = ["based on", "data", "forecast", "pricing", "solar", "usage"]
    if any(k in response_lower for k in data_keywords):
        score += 1
    else:
        feedback.append("Response does not clearly reference data or analysis.")

    # 4. Check for savings or optimization mention
    savings_keywords = ["save", "savings", "cost", "$", "reduce"]
    if any(k in response_lower for k in savings_keywords):
        score += 1
    else:
        feedback.append("Response does not quantify or mention cost savings.")

    # 5. Alignment with expected response intent
    if any(word in response_lower for word in expected_lower.split()):
        score += 1
    else:
        feedback.append("Response does not align well with expected intent.")

    return {
        "score": score,
        "max_score": 5,
        "feedback": feedback,
    }
```

```python
# TODO: Create a tool udage evaluator
def evaluate_tool_usage(messages, expected_tools):
    """
    Evaluate whether the agent used the expected tools.

    Args:
        messages: list of agent messages (LangChain format)
        expected_tools: list of tool names expected to be used

    Returns:
        dict with score and feedback
    """

    used_tools = set()
    feedback = []

    # Extract tool calls from messages
    for msg in messages:
        if isinstance(msg, dict):
            # OpenAI / LangChain tool call format
            if msg.get("type") == "tool":
                used_tools.add(msg.get("name"))
            if "tool_calls" in msg:
                for tool_call in msg["tool_calls"]:
                    used_tools.add(tool_call.get("name"))

    expected_tools = set(expected_tools)

    missing_tools = expected_tools - used_tools
    unexpected_tools = used_tools - expected_tools

    score = 0

    if not missing_tools:
        score += 2
    else:
        feedback.append(f"Missing expected tools: {list(missing_tools)}")

    if not unexpected_tools:
        score += 1
    else:
        feedback.append(f"Used unexpected tools: {list(unexpected_tools)}")

    if used_tools:
        score += 1
    else:
        feedback.append("No tools were used at all.")

    return {
        "score": score,
        "max_score": 4,
        "used_tools": list(used_tools),
        "missing_tools": list(missing_tools),
        "unexpected_tools": list(unexpected_tools),
        "feedback": feedback,
    }
```

```python
# TODO: Generate a comprehensive evaluation report
def generate_evaluation_report(evaluation_input: dict):
    question = evaluation_input["question"]
    final_response = evaluation_input["final_response"]
    expected_response = evaluation_input["expected_response"]
    messages = evaluation_input["messages"]
    expected_tools = evaluation_input["expected_tools"]

    response_eval = evaluate_response(
        question, final_response, expected_response
    )
    tool_eval = evaluate_tool_usage(messages, expected_tools)

    total_score = response_eval["score"] + tool_eval["score"]
    max_score = response_eval["max_score"] + tool_eval["max_score"]

    strengths, weaknesses, recommendations = [], [], []

    if response_eval["score"] >= 4:
        strengths.append("Strong, actionable, data-driven response.")
    else:
        weaknesses.extend(response_eval["feedback"])
        recommendations.append(
            "Improve response clarity, data grounding, and savings quantification."
        )

    if tool_eval["score"] >= 3:
        strengths.append("Appropriate and sufficient tool usage.")
    else:
        weaknesses.extend(tool_eval["feedback"])
        recommendations.append(
            "Ensure all required tools are called before making recommendations."
        )

    return {
        "overall_score": total_score,
        "max_score": max_score,
        "percentage": round(total_score / max_score * 100, 2),
        "strengths": strengths,
        "weaknesses": weaknesses,
        "recommendations": recommendations,
        "details": {
            "response_evaluation": response_eval,
            "tool_usage_evaluation": tool_eval,
        },
    }
```

```python
def evaluate_tool_usage(messages, expected_tools):
    """
    Evaluate whether the agent used the expected tools.

    Args:
        messages: list of agent messages (LangChain format)
        expected_tools: list of tool names expected to be used

    Returns:
        dict with score and feedback
    """
    
    used_tools = set()
    feedback = []

    # Extract tool calls from messages
    for msg in messages:
        # Check if it's a LangChain AIMessage with tool_calls
        if hasattr(msg, 'tool_calls') and msg.tool_calls:
            for tool_call in msg.tool_calls:
                used_tools.add(tool_call.get('name'))
        
        # Check if it's a LangChain ToolMessage
        elif hasattr(msg, 'name') and msg.name:
            used_tools.add(msg.name)
        
        # Check if it's a dictionary format
        elif isinstance(msg, dict):
            if msg.get('type') == 'tool':
                used_tools.add(msg.get('name'))
            elif 'tool_calls' in msg:
                for tool_call in msg['tool_calls']:
                    if 'function' in tool_call:
                        used_tools.add(tool_call['function'].get('name'))
                    elif 'name' in tool_call:
                        used_tools.add(tool_call.get('name'))

    expected_tools_set = set(expected_tools)
    missing_tools = expected_tools_set - used_tools
    unexpected_tools = used_tools - expected_tools_set

    score = 0

    if not missing_tools:
        score += 2
    else:
        feedback.append(f"Missing expected tools: {list(missing_tools)}")

    if not unexpected_tools:
        score += 1
    else:
        feedback.append(f"Used unexpected tools: {list(unexpected_tools)}")

    if used_tools:
        score += 1
    else:
        feedback.append("No tools were used at all.")

    return {
        "score": score,
        "max_score": 4,
        "used_tools": list(used_tools),
        "missing_tools": list(missing_tools),
        "unexpected_tools": list(unexpected_tools),
        "feedback": feedback,
    }
```

```python
all_reports = []

for i, test_result in enumerate(test_results):
    print(f"\n=== Processing Test {i+1}: {test_result['test_id']} ===")
    
    try:
        # Extract data from test result
        question = test_result['question']
        expected_response = test_result['expected_response']
        expected_tools = test_result['expected_tools']
        
        # Get messages from response
        if isinstance(test_result['response'], dict) and 'messages' in test_result['response']:
            messages = test_result['response']['messages']
            # Debug: print message types
            print(f"Message types: {[type(m).__name__ for m in messages]}")
            
            # Find the last AI message that has content
            for msg in reversed(messages):
                if hasattr(msg, 'content') and msg.content:
                    final_response = msg.content
                    break
                elif isinstance(msg, dict) and msg.get('content'):
                    final_response = msg.get('content')
                    break
            else:
                final_response = "No response content found"
        else:
            messages = []
            final_response = str(test_result['response'])
        
        # Debug: check for tool calls
        tool_calls_found = 0
        for msg in messages:
            if hasattr(msg, 'tool_calls') and msg.tool_calls:
                tool_calls_found += len(msg.tool_calls)
                print(f"Found {len(msg.tool_calls)} tool call(s) in {type(msg).__name__}")
                for tc in msg.tool_calls:
                    print(f"  - Tool: {tc.get('name')}")
        
        print(f"Total tool calls found: {tool_calls_found}")
        
        # Generate evaluation report
        response_eval = evaluate_response(
            question, final_response, expected_response
        )
        tool_eval = evaluate_tool_usage(messages, expected_tools)
        
        # Calculate overall score
        total_score = response_eval["score"] + tool_eval["score"]
        max_score = response_eval["max_score"] + tool_eval["max_score"]
        
        strengths, weaknesses, recommendations = [], [], []

        if response_eval["score"] >= 4:
            strengths.append("Strong, actionable, data-driven response.")
        else:
            weaknesses.extend(response_eval["feedback"])
            recommendations.append(
                "Improve response clarity, data grounding, and savings quantification."
            )

        if tool_eval["score"] >= 3:
            strengths.append("Appropriate and sufficient tool usage.")
        else:
            weaknesses.extend(tool_eval["feedback"])
            recommendations.append(
                "Ensure all required tools are called before making recommendations."
            )
        
        # Create report
        report = {
            "overall_score": total_score,
            "max_score": max_score,
            "percentage": round(total_score / max_score * 100, 2) if max_score > 0 else 0,
            "strengths": strengths,
            "weaknesses": weaknesses,
            "recommendations": recommendations,
            "details": {
                "response_evaluation": response_eval,
                "tool_usage_evaluation": tool_eval,
            },
        }
        
        # Store simplified report
        simplified_report = {
            "test_id": test_result['test_id'],
            "overall_score": report['overall_score'],
            "max_score": report['max_score'],
            "percentage": report['percentage'],
            "strengths": report['strengths'],
            "weaknesses": report['weaknesses'],
            "recommendations": report['recommendations'],
            "details": {
                "response_evaluation": report['details']['response_evaluation'],
                "tool_usage_evaluation": report['details']['tool_usage_evaluation'],
            }
        }
        
        all_reports.append(simplified_report)
        
        # Display minimal output
        print(f"Score: {simplified_report['overall_score']}/{simplified_report['max_score']} ({simplified_report['percentage']}%)")
        print(f"Tools used: {tool_eval['used_tools']}")
        print(f"Missing tools: {tool_eval['missing_tools']}")
        
    except Exception as e:
        print(f"Error processing test {test_result['test_id']}: {str(e)[:100]}")
        import traceback
        traceback.print_exc()
        error_report = {
            "test_id": test_result['test_id'],
            "error": str(e),
            "overall_score": 0,
            "max_score": 9,
            "percentage": 0,
            "strengths": [],
            "weaknesses": ["Failed to process test"],
            "recommendations": ["Check test data format"],
            "details": {}
        }
        all_reports.append(error_report)

# Display summary of all reports
print("\n" + "="*60)
print("SUMMARY OF ALL TEST RESULTS")
print("="*60)

successful_tests = [r for r in all_reports if 'error' not in r]
failed_tests = [r for r in all_reports if 'error' in r]

if successful_tests:
    print(f"Successful tests: {len(successful_tests)}")
    print("\nTest Results Summary:")
    for report in successful_tests:
        print(f"\n{report['test_id']}: {report['overall_score']}/{report['max_score']} ({report['percentage']}%)")
        
        if report['details'].get('tool_usage_evaluation', {}).get('used_tools'):
            tools = report['details']['tool_usage_evaluation']['used_tools']
            print(f"  Tools used: {len(tools)} tool(s): {tools}")
        
        if report['weaknesses']:
            print(f"  Issues: {len(report['weaknesses'])} issue(s)")

if failed_tests:
    print(f"\nFailed tests: {len(failed_tests)}")
    for report in failed_tests:
        print(f"\n{report['test_id']}: ERROR - {report.get('error', 'Unknown error')[:100]}")

# Calculate overall statistics
if successful_tests:
    total_score = sum(r['overall_score'] for r in successful_tests)
    total_max = sum(r['max_score'] for r in successful_tests)
    avg_percentage = sum(r['percentage'] for r in successful_tests) / len(successful_tests)
    
    print("\n" + "="*60)
    print("OVERALL STATISTICS")
    print("="*60)
    print(f"Total Tests: {len(successful_tests)}")
    print(f"Average Score: {total_score/len(successful_tests):.2f}/{total_max/len(successful_tests):.2f}")
    print(f"Average Percentage: {avg_percentage:.2f}%")
    print(f"Total Score: {total_score}/{total_max}")
    
    # Calculate grade distribution
    grade_distribution = {
        "A (90-100%)": len([r for r in successful_tests if r['percentage'] >= 90]),
        "B (80-89%)": len([r for r in successful_tests if 80 <= r['percentage'] < 90]),
        "C (70-79%)": len([r for r in successful_tests if 70 <= r['percentage'] < 80]),
        "D (60-69%)": len([r for r in successful_tests if 60 <= r['percentage'] < 70]),
        "F (<60%)": len([r for r in successful_tests if r['percentage'] < 60]),
    }
    
    print("\nGrade Distribution:")
    for grade, count in grade_distribution.items():
        percentage = (count / len(successful_tests)) * 100 if successful_tests else 0
        print(f"  {grade}: {count} test(s) ({percentage:.1f}%)")

# Display sample report for first test
if successful_tests:
    print("\n" + "="*60)
    print("SAMPLE DETAILED REPORT (First Test)")
    print("="*60)
    sample = successful_tests[0]
    print(f"Test ID: {sample['test_id']}")
    print(f"Overall Score: {sample['overall_score']}/{sample['max_score']} ({sample['percentage']}%)")
    
    resp_eval = sample['details']['response_evaluation']
    print(f"\nResponse Evaluation: {resp_eval['score']}/{resp_eval['max_score']}")
    if resp_eval.get('feedback'):
        print(f"Feedback: {resp_eval['feedback']}")
    
    tool_eval = sample['details']['tool_usage_evaluation']
    print(f"\nTool Usage: {tool_eval['score']}/{tool_eval['max_score']}")
    print(f"Used Tools: {tool_eval['used_tools']}")
    print(f"Missing Tools: {tool_eval['missing_tools']}")
    print(f"Unexpected Tools: {tool_eval['unexpected_tools']}")
```
